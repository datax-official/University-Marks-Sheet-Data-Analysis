# Filling other columns values "Numerical Columns"

Filling missing values with 0 in Numeric columns

```python
# Fill missing values in numerical columns with the mean
numerical_columns = ['Sessional(75)', 'Terminal(75)']
combined_df[numerical_columns] = combined_df[numerical_columns].fillna(0,inplace=False)
```

Filling with Mean

```python
# Specify the numerical columns
numerical_columns = ['Sessional(75)', 'Terminal(75)']

# Convert columns to numeric data types
combined_df[numerical_columns] = combined_df[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Fill missing values in numerical columns with the mean
combined_df[numerical_columns] = combined_df[numerical_columns].fillna(combined_df[numerical_columns].mean(),inplace=False)
```
